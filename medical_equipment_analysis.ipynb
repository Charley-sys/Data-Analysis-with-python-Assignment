This notebook demonstrates how to load, analyze, and visualize a **medical equipment dataset** using `pandas`, `matplotlib`, and `seaborn`.

---


```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# -------------------------
# 1. Load Dataset (simulate medical equipment dataset)
# -------------------------
try:
    data = {
        "Equipment_ID": [101, 102, 103, 104, 105, 106, 107, 108],
        "Equipment_Name": ["X-Ray Machine", "MRI Scanner", "Ventilator", "Ultrasound", 
                           "ECG Machine", "Defibrillator", "Patient Monitor", "Dialysis Machine"],
        "Department": ["Radiology", "Radiology", "ICU", "Radiology", 
                       "Cardiology", "Emergency", "ICU", "Nephrology"],
        "Quantity": [5, 2, 10, 7, 15, 8, 20, 4],
        "Cost_USD": [50000, 1500000, 20000, 30000, 5000, 10000, 7000, 40000],
        "Maintenance_Status": ["Good", "Needs Repair", "Good", "Good", 
                               "Out of Service", "Good", "Good", "Needs Repair"]
    }
    
    df = pd.DataFrame(data)
    print("Medical Equipment Dataset Loaded Successfully!\n")
    print("First 5 Rows:\n", df.head(), "\n")
    print("Dataset Info:\n")
    print(df.info(), "\n")
    print("Statistical Summary:\n", df.describe())
    
except Exception as e:
    print("Error loading dataset:", e)
```

```python
# -------------------------
# 2. Basic Data Analysis
# -------------------------
print("\nChecking for missing values:\n", df.isnull().sum(), "\n")
print("Maintenance Status Counts:\n", df['Maintenance_Status'].value_counts(), "\n")
print("Departments Covered:\n", df['Department'].unique())
```

```python
# -------------------------
# 3. Data Visualization
# -------------------------
sns.set(style="whitegrid")

# Bar chart: Quantity of Equipment by Department
plt.figure(figsize=(8,5))
sns.barplot(x="Department", y="Quantity", data=df, estimator=sum, palette="Set2")
plt.title("Total Equipment Quantity by Department")
plt.xlabel("Department")
plt.ylabel("Total Quantity")
plt.xticks(rotation=45)
plt.show()

# Pie chart: Maintenance Status
plt.figure(figsize=(6,6))
df['Maintenance_Status'].value_counts().plot.pie(autopct="%1.1f%%", colors=sns.color_palette("pastel"))
plt.title("Maintenance Status Distribution")
plt.ylabel("")
plt.show()

# Scatter Plot: Cost vs Quantity
plt.figure(figsize=(8,6))
sns.scatterplot(x="Cost_USD", y="Quantity", hue="Department", size="Quantity", data=df, palette="Set1", sizes=(50,300))
plt.title("Equipment Cost vs Quantity by Department")
plt.xlabel("Cost (USD)")
plt.ylabel("Quantity")
plt.show()
```

```python
# -------------------------
# 4. Findings / Observations
# -------------------------
print("\nKey Observations:")
print("- ICU has the highest number of patient monitors and ventilators, showing priority in critical care.")
print("- MRI scanner is the most expensive equipment but has low quantity (only 2).")
print("- Some equipment is marked as 'Needs Repair' or 'Out of Service', highlighting maintenance gaps.")
print("- Radiology department has high-value machines (X-Ray, MRI, Ultrasound).")
print("- Cost vs Quantity chart shows that low-cost equipment (ECG, monitors) are purchased in higher numbers.")
```
